```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

data=pd.read_csv("/content/olympics.csv")
data.head()
```

```python
# prompt: generate 10 different plots for the dataset for the columns:country	country_code	region	gold	silver	bronze	total	gdp	gdp_year	population

# Plot 1: Gold medals distribution by country
plt.figure(figsize=(12, 6))
sns.countplot(x='country', data=data, order=data['country'].value_counts().index[:10])
plt.title('Top 10 Countries by Number of Gold Medals')
plt.xticks(rotation=45)
plt.show()

# Plot 2: Total medals vs GDP scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp', y='total', data=data)
plt.title('Total Medals vs GDP')
plt.show()

# Plot 3: Distribution of total medals
plt.figure(figsize=(8, 5))
sns.histplot(data['total'], kde=True)
plt.title('Distribution of Total Medals')
plt.show()


# Plot 4: Box plot of GDP by region
plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='gdp', data=data)
plt.title('GDP Distribution by Region')
plt.xticks(rotation=45)
plt.show()


# Plot 5: Correlation heatmap for numerical columns
numerical_cols = ['gold', 'silver', 'bronze', 'total', 'gdp', 'population']
plt.figure(figsize=(10, 8))
sns.heatmap(data[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Plot 6: Bar plot of medals by region
plt.figure(figsize=(12,6))
data.groupby('region')['total'].sum().plot(kind='bar')
plt.title('Total Medals by Region')
plt.ylabel('Total Medals')
plt.xticks(rotation=45)
plt.show()


#Plot 7: Gold Medals vs Population
plt.figure(figsize=(10, 6))
sns.scatterplot(x='population', y='gold', data=data)
plt.title('Gold Medals vs Population')
plt.show()

#Plot 8: Silver Medals vs GDP
plt.figure(figsize=(10, 6))
sns.regplot(x='gdp', y='silver', data=data)
plt.title('Silver Medals vs GDP')
plt.show()

#Plot 9: Bronze Medals vs Year
plt.figure(figsize=(10,6))
sns.lineplot(x='gdp_year', y='bronze', data=data)
plt.title('Bronze Medals over Time')
plt.show()


# Plot 10: Pairplot for a subset of numerical columns
sns.pairplot(data[['gold', 'silver', 'bronze', 'total']])
plt.suptitle('Pairplot of Medal Counts')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Assuming 'total' is the target variable and other numerical columns are features
features = ['gold', 'silver', 'bronze', 'gdp', 'population']
target = 'total'

X = data[features]
y = data[target]

# Handle missing values (if any) - Imputation or removal
X = X.fillna(X.mean())  # Example: Fill with mean

# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

#Example prediction (requires a scaled input)

example_input = scaler.transform([[5,3,2,10000, 50000]]) #example scaled input
predicted_total_medals = model.predict(example_input)
print(f"Predicted Total Medals: {predicted_total_medals[0]}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Load your dataset (replace 'olympics.csv' with your actual file path)
df = pd.read_csv('/content/olympics.csv')

### 1. Data Preprocessing
# Fill missing values for numerical columns with median
df['gdp'] = df['gdp'].fillna(df['gdp'].median())
df['gdp_year'] = df['gdp_year'].fillna(df['gdp_year'].mode()[0])
df['population'] = df['population'].fillna(df['population'].mean())

# Fill missing values for categorical columns
df['region'] = df['region'].fillna('Unknown')

# Creating new features
df['gdp_per_capita'] = df['gdp'] / df['population']
df['medals_per_capita'] = df['total'] / df['population']
df['weighted_medal_score'] = df['gold'] * 3 + df['silver'] * 2 + df['bronze']

# Encoding categorical variables using one-hot encoding
df = pd.get_dummies(df, columns=['region'], drop_first=True)

# Drop unnecessary columns
df.drop(['country', 'country_code', 'gdp_year'], axis=1, inplace=True)

# Separate features (X) and target (y)
X = df.drop('total', axis=1)  # Features
y = df['total']  # Target: total number of medals

### 2. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

### 3. Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

### 4. Model Training with Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

### 5. Make Predictions
y_pred = model.predict(X_test)

### 6. Model Evaluation
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"R² Score: {r2:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

### 7. Feature Importance (Optional)
# Understanding which features are most important for predicting medal counts
feature_importances = model.feature_importances_
feature_names = X.columns

# Display the feature importances
important_features = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
important_features = important_features.sort_values(by='Importance', ascending=False)
print("\nFeature Importances:")
print(important_features)

### 8. Predicting Medals for India in the 2028 Olympics
# Assuming India's projected data for the 2028 Olympics
# Replace these values with realistic projections or estimates based on India's current trends
india_data = pd.DataFrame({
    'silver': [20],    # Example number of silver medals won in past Olympics
    'bronze': [22],    # Example number of bronze medals won in past Olympics
    'gold': [15],      # Example number of gold medals won in past Olympics
    'gdp': [3.5e12],   # India's projected GDP in 2028
    'population': [1.5e9],  # India's projected population in 2028
    'gdp_per_capita': [3.5e12 / 1.5e9],   # GDP per capita
    'medals_per_capita': [(15 + 20 + 22) / 1.5e9],  # Total medals / population
    'weighted_medal_score': [15 * 3 + 20 * 2 + 22]  # Weighted medal score
}, index=[0])

# One-hot encode the 'region' column using the same categories as in training
# Get all unique one-hot encoded 'region' columns from the original dataframe
region_columns = [col for col in df.columns if 'region_' in col]

# Initialize region columns with zeros
region_data = pd.DataFrame(np.zeros((1, len(region_columns))), columns=region_columns)

# Set the relevant region to 1 (e.g., 'region_Asia' for India)
region_data['region_Asia'] = 1  # Adjust this if your dataset has different regions

# Concatenate india_data with the region_data
sample_df = pd.concat([india_data, region_data], axis=1)

# Ensure that the sample_df has the same columns as the training set
missing_cols = set(X.columns) - set(sample_df.columns)
for col in missing_cols:
    sample_df[col] = 0

# Reorder columns to match the training data
sample_df = sample_df[X.columns]

# Scale the sample data
sample_scaled = scaler.transform(sample_df)

# Predict total medals for India in 2028
predicted_medals = model.predict(sample_scaled)
print(f"\nPredicted Total Medals for India in 2028: {int(predicted_medals[0])}")
```
